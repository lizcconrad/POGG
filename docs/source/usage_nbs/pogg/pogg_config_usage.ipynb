## Creating the config.yml file

First, let's create a `config.yml` using the ERG:

```yaml
# Grammar information
grammar_location: ./ERG_2023/erg-2023.dat
SEMI: ./ERG_2023/trunk/etc/erg.smi
```

## Initializing POGGConfig

Now, let's initialize a `POGGConfig` object:

````python
from pogg.pogg_config import POGGConfig

pogg_config = POGGConfig("config.yml")
````

For the most part, interacting with a `POGGConfig` object directly is not necessary. Instead, it will be sent as an argument when creating an instance of the `SemanticAlgebra` class, for instance.

Under the hood, the most important thing a `POGGConfig` object does is consult the SEM-I for information about the arguments of a semantic predicate and make that information "concrete."

### `concretize` Example

Let's look what information the SEM-I provides for a few predicates:

```
_cookie_n_1 : ARG0 x { IND + }.
_tasty_a_1 : ARG0 e, ARG1 u.
_eat_v_1 : ARG0 e, ARG1 i, ARG2 i.
```

The SEM-I tells us the following about each predicate:

1. The number of semantic arguments
2. The type of variable that can fill each argument
3. Potentially, variable properties for the argument

But, in order to work with actual semantic structures, the variables must be made "concrete." That is, the `ARG0` of `_cookie_n_1` can't just be `x`, as that is just a variable type, but rather `x1`, which is a specific instance of a variable of type `x`. The instance method `concretize` on a `POGGConfig` object does this for us:

````python
pogg_config.concretize("_cookie_n_1")
````

````python
pogg_config.concretize("_tasty_a_1")
````

````python
pogg_config.concretize("_eat_v_1")
````
