The SEMENT class is an extension of the [MRS class](https://pydelphin.readthedocs.io/en/latest/api/delphin.mrs.html) from [PyDelphin](https://pydelphin.readthedocs.io/en/latest/guides/setup.html).

The conceptual difference between a SEMENT and an MRS is that a SEMENT is a structure designed to enable semantic composition, while an MRS is considered to be a "finished" semantic structure.

As a result, a SEMENT has additional machinery to enable composition.

The formal details are described on the [SEMENT education page](project:/education/sement_education.ipynb).

Below is an example of a SEMENT in string format:

```
[
    TOP: h0
    INDEX: x1
    RELS: < [ _cookie_n_1 LBL: h0 ARG0: x1 ] >
]
```

To create a SEMENT object directly that would match the string shown above you would do the following:

````python
from pogg.my_delphin.my_delphin import SEMENT
from delphin.mrs import EP

# make elementary predicate for the RELS list
cookie_EP = EP("_cookie_n_1", "h0", {"ARG0": "x1"})

cookie_sement = SEMENT("h0", "x1", [cookie_EP])
````

Then using the [sementcodecs](project:/apidocs/pogg/pogg.my_delphin.sementcodecs.md) we can convert the object to a string to confirm it matches the example:

````python
import pogg.my_delphin.sementcodecs as sementcodecs

print(sementcodecs.encode(cookie_sement, indent=True))
````

While it would be possible to initialize a more complex SEMENT with multiple elements in the RELS list, various variable equalities, and handle constraints by hand, it would be very cumbersome. Instead, the [SemanticAlgebra class](project:/apidocs/pogg/pogg.semantic_composition.semantic_algebra.md) is used to create "starter" SEMENTs (much like the one above) and then perform composition with various functions.
