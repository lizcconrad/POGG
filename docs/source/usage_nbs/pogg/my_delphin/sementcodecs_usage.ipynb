# sementcodecs usage example

[See full module documentation here.](project:/apidocs/pogg/pogg.my_delphin.sementcodecs.md)

The `sementcodecs` module includes functions to convert SEMENT objects to strings, and vice versa.

## Decode string into SEMENT object

```python
import pogg.my_delphin.sementcodecs as sementcodecs

cookie_SEMENT_str = """
[
    TOP: h0
    INDEX: x1
    RELS: < [ _cookie_n_1 LBL: h0 ARG0: x1 ] >
]
"""

# decode string into SEMENT object
cookie_SEMENT_obj = sementcodecs.decode(cookie_SEMENT_str)

# print properties of the SEMENT object
print(cookie_SEMENT_obj)
print(cookie_SEMENT_obj.top)
print(cookie_SEMENT_obj.index)
print(cookie_SEMENT_obj.rels)
```

## Encode SEMENT object into string

Let's modify the SEMENT by hand, changing the TOP handle to `h2`. This will create a broken SEMENT, but it will illustrate that our changes are reflected in the encoded string once we print it.

```python
cookie_SEMENT_obj.top = "h2"

# encode the SEMENT object into string and print
print(sementcodecs.encode(cookie_SEMENT_obj, indent=True))
```
