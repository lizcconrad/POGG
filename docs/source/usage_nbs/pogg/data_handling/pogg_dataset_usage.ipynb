A `POGGDataset` object stores information about a provided dataset such as where it's located and where any output should be directed.

## Example dataset structure

A POGG dataset should have the structure shown below.

For more information on the specifics of what goes into a POGG dataset, see the [What's in a POGG dataset?]() page. For more information on how to create a dataset for POGG, see the [Creating a dataset for POGG]() page.

```text
dataset_dir/
    - PetitePatisserie/     <-- each of these is a "chunk" of the dataset
    - LittleLibrary/
    - BitsyBakery/          <-- each chunk will have the subdirectories shown here
        - graph_jsons/      <-- graphs in JSON format
            - cake_graph.json
            - cookie_graph.json
        - dot/              <-- graphs in DOT format
            - cake_graph.dot
            - cookie_graph.dot
        - lexicon/          <-- lexicon files
            - lexicon_all_entries.json
            - lexicon_complete_entries.json
            - lexicon_incomplete_entries.json
            - lexicon_invalid_entries.json
        - evaluation/       <-- evaluation output from POGG algorithm will be stored here
        - gold_outputs/     <-- gold outputs for each graph are stored here
```

## Creating a `dataset_config.yml`

Assuming you have a dataset that follows the structure above, you can fill out a YAML config file with the appropriate information.

Below is an example of such a config file with all the necessary fields filled out:

```yaml
# dataset.yml example
# data information
dataset_name: "BitsyBakery"

# top level directory
data_dir: "/path_to_dataset/dataset_dir"

# subdirectories
data_chunk: "BitsyBakery"
graph_json_dir: "graph_jsons"
graph_dot_dir: "dot"
lexicon_dir: "lexicon"
evaluation_dir: "evaluation"
gold_outputs_dir: "gold_outputs"
```

The following tabe describes each field that must be included in the dataset's YAML config file.


**Config File Fields**
| Field | Description | Optional? |
| ----- | ----------- | ----- |
| `dataset_name` | Name of the dataset (or dataset chunk, if applicable) | No |
| `data_dir` | Top level directory storing the dataset | No |
| `data_chunk` | Subdirectory containing the relevant data chunk if the dataset is divided into chunks | Yes |
| `graph_json_dir` | Subdirectory containing graph data in JSON format | No |
| `graph_dot_dir` | Subdirectory containing graph data in DOT format | No |
| `lexicon_dir` | Subdirectory containing lexicon information | No |
| `evaluation_dir` | Subdirectory where evaluation results will be stored | No |
| `gold_outputs_dir` | Subdirectory where evaluation results will be stored | No |


## Initializing a POGGDataset object

Once you have your YAML config file, you can pass it to the `POGGDataset` initializer to get a `POGGDataset` object:



````python
from pogg.data_handling.pogg_dataset import POGGDataset

# create the POGGDataset object
bitsy_bakery_dataset = POGGDataset("dataset_config.yml")

# print some information about the POGGDataset object
print(bitsy_bakery_dataset)
print(bitsy_bakery_dataset.data_dir)
print(bitsy_bakery_dataset.dataset_name)
print(bitsy_bakery_dataset.lexicon_dir)
````
