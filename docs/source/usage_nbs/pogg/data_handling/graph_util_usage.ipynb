# POGGGraphUtil usage example

[See full module documentation here.](project:/apidocs/pogg/pogg.data_handling.graph_util.md)

The `POGGGraphUtil` class contains a number of static functions to make working with directed graphs within POGG easier.

## `build_graph` example

The static function `build_graph` takes in a JSON object containing information about a graph and builds a NetworkX directed graph out of it. See the [Converting graphs to JSON format]() page for more details on the specific format this function requires.

:::{info} Function details
:collapsible:
````{py:function} build_graph(graph_json)
:canonical: pogg.data_handling.graph_util.POGGGraphUtil.build_graph

```{autodoc2-docstring} pogg.data_handling.graph_util.POGGGraphUtil.build_graph
```

````
:::


Let's say we have a JSON file with the following information:

```json
    {
        "nodes": {
            "idCar1": {
                "lexicon_key": "idCar",
                "node_properties": {
                    "node_type": "entity",
                    "root": "root"
                }
            },
            "red": {
                "node_properties": {
                    "node_type": "property"
                }
            }
        },
        "edges": [
            {
                "edge_name": "idColor",
                "parent_node": "idCar1",
                "child_node": "red",
                "lexicon_key": "idColor",
                "edge_properties": {
                    "edge_type": "property"
                }
            }
        ]
    }
```

The graph this represents is a small graph with just two nodes and one edge:

<img src="../../../data/images/build_graph_example.png" alt="Directed graph with a parent node labeled 'idCar1' with an edge labeled 'idColor' pointing to a child node labeled 'red'" width="20%" height="auto">

We can load this JSON file with the `json` library and then use the `build_graph`  function in `POGGGraphUtil` to create a NetworkX directed graph:

`````python
import json
from pogg.data_handling.graph_util import POGGGraphUtil

graph_json = json.load(open("graph.json"))

graph = POGGGraphUtil.build_graph(graph_json)

# confirm we got a NetworkX DiGraph object
print(graph)
`````

## `find_root` example

The recursive graph-to-SEMENT algorithm POGG uses relies on finding the root of the provided graph to start. This function helps to do that. Ideally, one of the nodes in the graph will be marked as the root, but if it isn't then the function will try a few more things to figure it out, ultimately returning `None` if it fails to find a suitable unique candidate for the root.

:::{info} Function details
:collapsible:
````{py:function} find_root(graph)
:canonical: pogg.data_handling.graph_util.POGGGraphUtil.find_root

```{autodoc2-docstring} pogg.data_handling.graph_util.POGGGraphUtil.find_root
```

````
:::

For the graph we created in the above example, the root should be the node labeled `idCar1`. In addition to the name of the node it will also return all of that node's properties:

`````python
root = POGGGraphUtil.find_root(graph)
print(root)
`````

## Read/Write Examples

The remaining functions in POGGGraphUtil are functions used to either read from existing DOT files or write graphs to various file formats.

`````python
# read from an existing DOT file
existing_graph = POGGGraphUtil.read_graph_from_dot("existing_graph.dot")
print(existing_graph)
`````

`````python
# write the graph from the previous examples to various file formats
POGGGraphUtil.write_graph_to_dot(graph, "graph.dot")
POGGGraphUtil.write_graph_to_png(graph, "graph.png")
POGGGraphUtil.write_graph_to_svg(graph, "graph.svg")
`````
