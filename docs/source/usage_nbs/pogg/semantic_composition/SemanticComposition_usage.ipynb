# `SemanticComposition` usage examples

This page contains examples for using the `SemanticComposition` class.

[See full module documentation here](project:/apidocs/pogg/pogg.semantic_composition.semantic_composition.md)

The `SemanticComposition` class inherits from a number of "mixin" classes that each have a set of semantic composition functions. Because the number of composition functions is vast and grows over time, it is helpful to organize them into smaller classes, and then have a class that inherits from all existing Mixins whose instances can call any of the composition functions when needed.

This guide will show an example for each composition function, organized by the Mixins that define them.

First, we will need the following imports:

`````python
import pogg.my_delphin.sementcodecs as sementcodecs
from pogg.pogg_config import POGGConfig
from pogg.semantic_composition.semantic_algebra import SemanticAlgebra
from pogg.semantic_composition.semantic_composition import SemanticComposition
`````

Next, we'll assume there is a `config.yml` file with the following information:

```yaml
# Grammar information
grammar_location: ./ERG_2023/erg-2023.dat
SEMI: ./ERG_2023/trunk/etc/erg.smi
```

Then we'll use this file to initialize a `POGGConfig` object and a `SemanticAlgebra` object and use that to initialize a `SemanticComposition` object:

`````python
pogg_config = POGGConfig("../config.yml")
semantic_algebra = SemanticAlgebra(pogg_config)
semantic_composition = SemanticComposition(semantic_algebra)
`````

## `SingleWordConstructionsMixin`

The `SingleWordConstructionsMixin` contains functions that produce "starter" SEMENTs. These typically have only one predication in the RELS (relations) list and map to roughly one word of English. Almost all the functions are just a wrapper around the `create_base_SEMENT` function in the `SemanticAlgebra` class, but their names are more transparent for users.

### `create_base_SEMENT` wrappers

The following functions are simple wrappers around the `create_base_SEMENT` function from the `SemanticAlgebra` class:

- `basic` (used when part of speech is not known, for example)
- `adjective`
- `determiner`
- `noun`
- `preposition`
- `quantifier`
- `verb`

Each of these functions takes a predicate label and an optional dictionary of intrinsic variable properties as arguments. In reality, these functions all do the same thing under the hood: pass the arguments directly to `create_base_SEMENT`, but the idea behind having wrappers named after parts of speech is to make the semantic composition class more intuitive for users.


`````python
# some example SEMENTs
eat = semantic_composition.verb("_eat_v_1", {'TENSE': 'past'})
the = semantic_composition.determiner("_the_q") # could also use quantifier
tasty = semantic_composition.adjective("_tasty_a_1")
cookie = semantic_composition.noun("_cookie_n_1", {'NUM': 'sg'})


# print the SEMENTs as a strings
print(sementcodecs.encode(eat, indent=True))
print(sementcodecs.encode(the, indent=True))
print(sementcodecs.encode(tasty, indent=True))
print(sementcodecs.encode(cookie, indent=True))
`````

This function generates a starter SEMENT for a named entity. It's a simple wrapper around `create_CARG_SEMENT` in the `SemanticAlgebra` class. It takes a string for the named entity (e.g. "Liz") and a dictionary of intrinsic variable properties as arguments.


`````python
liz = semantic_composition.named_entity("Liz", {'NUM': 'sg'})

# print the SEMENT as a string
print(sementcodecs.encode(the, indent=True))
`````
