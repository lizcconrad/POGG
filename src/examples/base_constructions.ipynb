```python
from pogg.pogg_config import POGGCompositionConfig
import pogg.my_delphin.sementcodecs as sementcodecs
import pogg.semantic_composition.base_constructions as base_constructions

from delphin import ace
from delphin.codecs import simplemrs
import pogg.semantic_composition.semantic_algebra as semantic_algebra

pogg_config = POGGCompositionConfig("../data/config.yml")


def generate_from_sement(pogg_config, pre_sement):
    sement = semantic_algebra.prepare_for_generation(pogg_config, pre_sement)

    if type(sement) is str:
        # decode into SEMENT object if given as a string
        sement = sementcodecs.decode(sement)

    # encode into MRS string, since that's the format the ERG needs
    mrs_string = simplemrs.encode(sement, indent=True)

    with ace.ACEGenerator(pogg_config.grammar_location, ['-r', 'root_frag']) as generator:
        print("\n\nGENERATING FROM ... ")
        print("{}\n".format(mrs_string))
        generator_response = generator.interact(mrs_string)
        print("GENERATED RESULTS ... ")
        for r in generator_response.results():
            print(r.get('surface'))
        print("\n\n")

```
